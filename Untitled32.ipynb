1.> What is hypothesis testing in statistics
Hypothesis testing is a structured process that allows researchers to test assumptions (hypotheses) about a population parameter. It involves collecting sample data and determining whether the evidence supports rejecting a stated assumption. This method is widely used in fields like medicine, psychology, economics, and engineering.

```python
2.What is the null hypothesis, and how does it differ from the alternative hypothesis
Null Hypothesis (H₀): This is the default assumption that there is no effect or difference. For example, "The average test score is 70."

Alternative Hypothesis (H₁ or Ha): This posits that there is an effect or difference. For example, "The average test score is not 70."
```

```python
3.What is the significance level in hypothesis testing, and why is it important
The significance level, denoted by α, is the threshold for determining whether a result is statistically significant. Commonly set at 0.05, it represents a 5% risk of rejecting the null hypothesis when it is actually true (Type I error).

```

4. What does a P-value represent in hypothesis testing
The p-value indicates the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true. A smaller p-value suggests stronger evidence against the null hypothesis. If the p-value is less than or equal to α, the null hypothesis is rejected.

```python
5.How do you interpret the P-value in hypothesis testing
Low p-value (≤ α): Strong evidence against H₀; reject H₀.

High p-value (> α): Insufficient evidence against H₀; fail to reject H₀.

It's important to note that failing to reject H₀ does not prove it true; it merely indicates a lack of evidence against it.
```

```python
6.> What are Type 1 and Type 2 errors in hypothesis testing
Type I Error (False Positive): Rejecting H₀ when it is actually true.

Type II Error (False Negative): Failing to reject H₀ when it is actually false.

Balancing these errors is crucial in hypothesis testing.
```

```python
7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing
One-Tailed Test: Tests for an effect in one direction (e.g., greater than).

Two-Tailed Test: Tests for an effect in both directions (e.g., not equal to).

The choice depends on the research question and hypothesis.
```

```python
8.What is the Z-test, and when is it used in hypothesis testing
Z-Test: Used when the population standard deviation is known and the sample size is large (n ≥ 30).

T-Test: Used when the population standard deviation is unknown and/or the sample size is small (n < 30).
```

```python
9.How do you calculate the Z-score, and what does it represent in hypothesis testing
Confidence Interval: A range of values within which the true population parameter is expected to lie with a certain level of confidence (e.g., 95%).

Margin of Error: The maximum expected difference between the true population parameter and a sample estimate.
```

```python
4.E Perform a two-tailed Z-test using Python and visualize the decision region on a plotE
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test_two_tailed(sample_mean, pop_mean, std_dev, n, alpha=0.05):
    z = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))

    # Critical z value for two-tailed test
    z_crit = stats.norm.ppf(1 - alpha / 2)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')
    
    # Rejection regions
    plt.fill_between(x, y, where=(x < -z_crit) | (x > z_crit), color='red', alpha=0.3, label='Rejection Region')
    
    # Test statistic line
    plt.axvline(z, color='blue', linestyle='--', label=f'Z-statistic = {z:.2f}')
    
    plt.title('Two-tailed Z-test Decision Region')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

    return z, p

# Example usage
z_stat, p_val = z_test_two_tailed(sample_mean=508, pop_mean=500, std_dev=15, n=36)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
```

```python
5.E Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors for hypothesis testing.
    
    Parameters:
    - mu0: Mean under the null hypothesis
    - mu1: Mean under the alternative hypothesis
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (default = 0.05)
    """
    se = sigma / np.sqrt(n)
    z_crit = norm.ppf(1 - alpha)
    critical_value = mu0 + z_crit * se
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    h0_dist = norm.pdf(x, mu0, se)
    h1_dist = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_dist, label='H₀: μ = μ₀', color='blue')
    plt.plot(x, h1_dist, label='H₁: μ = μ₁', color='green')

    # Type I Error (False Positive)
    plt.fill_between(x, h0_dist, where=(x > critical_value), color='blue', alpha=0.3, label='Type I Error (α)')

    # Type II Error (False Negative)
    plt.fill_between(x, h1_dist, where=(x < critical_value), color='green', alpha=0.3, label='Type II Error (β)')

    # Vertical line for critical value
    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

    # Approximate beta (Type II error) using CDF under H1
    beta = norm.cdf(critical_value, loc=mu1, scale=se)
    power = 1 - beta

    return {"Critical Value": critical_value, "Type II Error (β)": beta, "Power (1-β)": power}
📌 Usage Example
python
Copy
Edit
results = visualize_type1_type2(
    mu0=100,     # Null hypothesis mean
    mu1=105,     # True mean under alternative
    sigma=10,    # Population standard deviation
    n=30,        # Sample size
    alpha=0.05   # Significance level
)

print(results)
```

```python
6.Write a Python program to perform an independent T-test and interpret the results
```

```python
import numpy as np
from scipy import stats

# Sample data for two independent groups
group1 = np.array([23, 21, 19, 24, 30, 22, 20, 18])
group2 = np.array([31, 35, 29, 32, 30, 28, 34, 33])

# Perform the independent t-test (assumes equal variances by default)
t_stat, p_value = stats.ttest_ind(group1, group2)

# Output the results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference between group means).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference between group means).")
```

```python
9.Write a Python function to calculate the confidence interval for a sample mean and explain its significanceE
```

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    h = se * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return mean - h, mean + h

data = np.random.normal(loc=100, scale=15, size=40)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

```python
def margin_of_error(data, confidence=0.95):
    se = stats.sem(data)
    z = stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return z * se

moe = margin_of_error(data)
print(f"Margin of Error (95%): {moe:.4f}")
```

```python
11.E Implement a Bayesian inference method using Bayes' Theorem in Python and explain the processE
```

```python
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: Disease test
p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05

posterior = bayes_theorem(p_disease, p_pos_given_disease, p_pos_given_no_disease)
print(f"Probability of disease given positive test: {posterior:.4f}")
```
